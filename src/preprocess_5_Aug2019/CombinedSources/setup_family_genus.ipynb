```python
import pandas as pd
from pprint import pprint
import json

file = 'combined_species.csv'
df = pd.read_csv(file,index_col=0)

df.head(10)
```

```python
family_genus_dict = {}

for i,row in df.iterrows():
    if type(row['family']) == str :
        family = row['family']
        sc_name = row['sc_name']
        gn = sc_name.split(' ')[0]
#         print (family,gn)
        if family in family_genus_dict.keys():
            family_genus_dict[family].append(gn)
        else :
            family_genus_dict[family] = [gn]

tmp_dict = {}
for k,v in family_genus_dict.items():
    tmp_dict[k] = list(set(v))

family_genus_dict = tmp_dict

pprint(family_genus_dict)
```

```python

filename = 'family_genus.json'
with open(filename, 'w') as outfile:
    data_str =  json.dumps(family_genus_dict, indent=4)
    outfile.write(data_str)
```
