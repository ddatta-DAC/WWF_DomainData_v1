```python
import pandas as pd
import textacy
from pprint import pprint
df = pd.read_excel('ForestKeyword_proc1.xlsx')
```

```python
def dumb_replace(sent):
    phrases = [ 
    'Central and South America'
    ]
    
    for ph in phrases :
        sent = sent.replace(ph)

    return sent
    
all_countries = []
for i,row in df.iterrows():
    cnt = row[2]
    if type(cnt)!= str :
        continue
    cnt = cnt.replace('\xa0','')
    cnt = cnt.split('\n')
    tmp = []
    for p in cnt :
        
        if p == '' :
            continue;
        print ('>> ',p)
        p = p.replace(' and',',')
        p_parts = p.split(',')
        for t in p_parts:
            tmp.append(t)
    all_countries.extend(tmp)
    
all_countries = list(set(all_countries))
pprint(all_countries)
```
