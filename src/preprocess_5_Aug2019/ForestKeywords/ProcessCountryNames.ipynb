```python
import pandas as pd
import textacy
from pprint import pprint
df = pd.read_excel('ForestKeyword_proc1.xlsx')
```

```python
def dumb_replace(sent):
    phrases = { 
    'Central and South America' : 'Central America , South America',
    'Dem. Rep. of the Congo' : 'Congo' ,
    'Dem. Rep of the Congo' : 'Congo',
    'Dem. Rep. Of the Congo' : 'Congo',
    'Trinidad and Tobago': 'Trinidad-Tobago'
    }
    
    for ph,rep in phrases.items():
        sent = sent.replace(ph,rep)

    return sent
```

```python
all_countries = []
for i,row in df.iterrows():
    cnt = row[2]
    if type(cnt)!= str :
        continue
    
    cnt = cnt.replace('\xa0','')
    cnt = dumb_replace(cnt)
    cnt = cnt.split('\n')
    tmp = []
    for p in cnt :
        if p == '' :
            continue;
        p = p.replace(' and',',')
        p_parts = p.split(',')
        for t in p_parts:
            t = t.strip()
            tmp.append(t)
    all_countries.extend(tmp)
    
all_countries = list(set(all_countries))
# pprint(all_countries)
```
